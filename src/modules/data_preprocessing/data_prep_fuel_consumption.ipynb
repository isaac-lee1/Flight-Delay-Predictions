```python
import pandas as pd
import numpy as np
```

```python
#import data into dataframe
df_fuel_raw = pd.read_csv('data/raw_data/fuel_consumption(raw).csv')
df_fuel_raw.head()
```

```python
#Let dataframe show all columns
pd.set_option('display.max_columns', None)
```

```python
#drop unnecessary columns
df_fuel = df_fuel_raw[['month', 'airline_id', 'carrier', 'carrier_name', 'total_gallons', 'total_cost', 'year']]
df_fuel.head()
```

```python
#Check for missing data
def check_null(df_fuel):
    total = df_fuel.isnull().sum().sort_values(ascending=False)
    percent = ((df_fuel.isnull().sum()/df_fuel.isnull().count()).sort_values(ascending=False))*100
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return missing_data
```

```python
check_null(df_fuel)
```

```python
#drop any rows with null values 
df_fuel_clean = df_fuel.dropna(how='any',axis=0) 
```

```python
check_null(df_fuel_clean)
```

```python
#compare dimensions of dataframes
print(df_fuel.shape)
print(df_fuel_clean.shape)
```

```python
#export dataframe as a csv
df_fuel_clean.to_csv('df_fuel_clean.csv', index=False)
```

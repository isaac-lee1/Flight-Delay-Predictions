```python
import pandas as pd
import numpy as np
```

```python
#import data into dataframe
df_passengers_raw = pd.read_csv('passengers(raw).csv')
df_passengers_raw.head()
```

```python
#Let dataframe show all columns
pd.set_option('display.max_columns', None)
```

```python
#drop unnecessary columns
df_passengers = df_passengers_raw[['departures_performed', 'seats', 'passengers', 'distance', 'air_time', 'unique_carrier', 'airline_id',
                         'origin_airport_id', 'origin', 'dest_airport_id', 'dest', 'aircraft_group', 'aircraft_type', 'year', 'month']]
df_passengers.head()
```

```python
#Check for missing data
def check_null(df_passengers):
    total = df_passengers.isnull().sum().sort_values(ascending=False)
    percent = ((df_passengers.isnull().sum()/df_passengers.isnull().count()).sort_values(ascending=False))*100
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return missing_data
```

```python
check_null(df_passengers)
```

```python
#export dataframe as a csv
df_passengers.to_csv('df_passengers_clean.csv', index=False)
```
